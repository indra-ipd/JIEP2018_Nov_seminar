## Mark Down による説明の記載

```python
import matplotlib.pyplot as plt

month = [1, 2, 3, 4, 5,6]
income = [100, 400, 200, 500, 700, 650]

#グラフを書く
plt.plot(month,income)

#グラフのタイトル
plt.title("effectiveness")

# x軸のラベル
plt.xlabel("month")

# y軸ラベル
plt.ylabel("income")

#表示する
plt.show()
```
